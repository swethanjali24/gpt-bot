```python
# Size Comparison
import sys
import numpy as np

# Declaring a list of 1000 elements
S =[1,2,3,4,5]

# Output
print('Size of each element of list in bytes :',[sys.getsizeof(item) for item in S])
print('Size of the whole list in bytes :', sys.getsizeof(S) * len(S))

# Declaring a numpy array
D = np.array([1,2,3,4,5])

# Output
print('Size of each element of the numpy array in bytes :', D.itemsize)
print('Size of the whole numpy array in bytes :', D.size * D.itemsize)
```

```python
# Dot and cross prodcuts
a=np.random.randint(low=1,high=5,size=(3,3))
print(a)
b=np.random.randint(low=1,high=5,size=(3,3))
print(b)
Dot_product=np.dot(a,b)
print(Dot_product)
Cross_product=np.cross(a,b)
print(Cross_product)# Row wise cross products as cross is generally for 2 vectors

print(Dot_product.reshape(1,-1))
print(Dot_product.reshape(-1,1))

print(Dot_product.reshape(1,-1).shape)
```

```python
# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Creating a scatter plot
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
plt.scatter(x, y,c='Green',label='datapoints')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()

```

```python
# Line plot
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
plt.figure(figsize=(5,3))
plt.plot(x, y,c='Red',label='datapoints')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
# Bar chart
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y,color='Green',label='frequency',width=0.2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Chart")
plt.legend()
plt.figure(figsize=(5,3))
plt.show()
```
